# K means cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters =2)
```

```python
kmean.fit(data.drop('Private',axis =1))
```

```python
kmean.cluster_centers_
```

```python
x = data.drop('Private', axis=1)
cluster1 = kmean.predict(x)
```

```python
data['cluster1'] = cluster1
```

```python
len(data[data['cluster1']==0])
```

```python
data.drop('cluster1', axis = 1, inplace = True)
```

```python
data.head()
```

```python
def converter(cluster):
    if cluster =='Yes':
        return 1
    else:
        return 0
```

```python
data['cluster'] = data['Private'].apply(converter)
```

```python
data.head()
```

```python
len(data[data['cluster']==1])
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(data['cluster'], kmean.labels_))
```

```python
print(confusion_matrix(data['cluster'], kmean.labels_))
```
